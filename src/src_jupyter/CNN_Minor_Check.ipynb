```python
import torch
from src.CNN_Training_Data_Load import train_loader
from src.CNN_Model_Device import model

x_cpu, y_cpu = next(iter(train_loader))  # 先看下数据格式，数据数字类型
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
x, y = x_cpu.to(device), y_cpu.to(device)
```
